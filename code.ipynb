```python
from main import Graph, dijsktra
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
nodes = {
    0:"Node A", 1:"Node B", 2:"Node C", 3:"Node D", 4:"Node G",
    5:"Node H", 6:"Node I", 7:"Node K", 8:"Node L", 9:"Node M"
    }

repeaters = {
    10:"repeater E", 11:"repeater F", 12:"repeater J"
    }

G.add_nodes_from(nodes)
G.add_nodes_from(repeaters)

num_weight = 20
G.add_edge(0, 10, weight=num_weight)
G.add_edge(1, 10, weight=num_weight)
G.add_edge(2, 10, weight=num_weight)
G.add_edge(3, 10, weight=num_weight)
G.add_edge(4, 11, weight=num_weight)
G.add_edge(5, 11, weight=num_weight)
G.add_edge(6, 11, weight=num_weight)
G.add_edge(7, 12, weight=num_weight)
G.add_edge(8, 12, weight=num_weight)
G.add_edge(9, 12, weight=num_weight)

G.add_edge(10, 11, weight=num_weight)
G.add_edge(11, 12, weight=num_weight)

H = nx.relabel_nodes(G, nodes)
H = nx.relabel_nodes(H, repeaters)

nx.draw(H, with_labels=True)
plt.show()
```
